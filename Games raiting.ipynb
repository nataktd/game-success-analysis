*Посмотрим, как менялись продажи по платформам, выберем платформы с наибольшими суммарными продажами и постройм распределение по годам. Определим за какой характерный срок появляются новые и исчезают старые платформы.*

*Определим актуальный период на основе результатов предыдущего исследования.*

*Основным фактором для определения актуального периода — эти данные помогут построить прогноз на 2017 год и более годы.*

*Выведем 15 строк сводной таблица. Как распрелелился рейтинг ведуших платформ рассмотрим на графике. В таблице же хотим увидеть , а зависят ли продажи от количества выпущенных на платформе игр. Помимо данных о продажах по каждой платформе (в количественном и процентном соотношении от общего объема продаж) выведем данные о количестве выпущенных на этой платформе игр.*

```python
def percent_category(new_df, index, column):
    try:
        sort_df = (new_df.pivot_table(index=index,
                                      values=column, 
                                      aggfunc=['count','sum']))        
        sort_df['percent'] = (sort_df['sum'] / sort_df['sum'].sum()) * 100 
                
        sort_df.columns = ['Количество выпущенных игр','Сумма продаж (в млн шт)',  'Процент от общего по продажам']
        return sort_df
    except:
        return 'Не верно введен аргумент'
    
platform_pivot = percent_category(games, 'platform', 'total_sales') # вызовем метод
platform_pivot.sort_values(by='Количество выпущенных игр', ascending = False).head(15)
```

В продажах платформа `DS` стоит на 5 месте.Например платформа `X360` занимает 5 место по количеству игр, но в рейтинге популярных (по продажам) платформ занимет 2 место.*

*Построим диаграмму рассеивания для платформы `PS4` и отобразим на ней корреляцию между отзывами и продажами.*

*Обьявим универсальный метод `dependence_price_feature` для диаграммы рассеяния и кореляционной прямой.*

```python
def dependence_sales_score(new_df, column, feature):
    platform_corr = new_df.reset_index() # сводная таблица#
        
    # график 
    platform_corr.plot(y=column, x='total_sales', kind='scatter', grid=True, figsize=(15, 5), color='C5', alpha=0.7)
        
    plt.xlabel('Продажи')
    plt.ylabel(feature)
    
    y = platform_corr[column]
    x = platform_corr['total_sales']
    plt.suptitle('Диаграмма рассеяния. Зависимость продаж по платформе от оценок.')
    
    m, b = np.polyfit(x, y, 1)
                 
    plt.plot(x, m*x + b, color='red') # линия регрессии
    plt.show()
    
    print('Коэффициент корреляции по платформе:', 
          platform_corr['total_sales'].corr(platform_corr[column]).round(2)) 
    print('Средняя оценка по платформе:', 
          platform_corr[column].mean().round(2)) 
    
```

```python
dependence_sales_score(ps4_platform, 'critic_score', 'Оценки критиков') # вызовем метод
```

Они покажут нам процент изменений (рос или падение) и в свою очередь смогут дать писимистический (оптимистеческий) прогноз на следующий год и позволят составить рейтинг ТОП жанров, на которые следую делать упор в продажах в будущем.*

*Расчитаем относительный показатель изменения количества выпускаемых на рынок новых игр, как соотношение выпуска в 2015 году к выпуску в 2014 году ( проверка с данными 2013 года по итогу всех расчетов дала тот же результат). Если полученный коеффициент будет меньше 1, значит наблюдаем спад выпуска, если выше единицы - рост выпуска новой продукции.*

*Такой же расчет счет сделаем и для медианного значения продаж.*

*Добавим в таблицу столбец по общим продажам за 2015 год.*

*С помощью полученных относительных показателей предскажем возможные общие продажи на следующий год исходя из динамики за последние три года и выделим ТОП жанров, которые дадут максимальные продажи в 2017 году.*

## **Портрет пользователя в разрезе регионов.**

*Определим для пользователя каждого региона (NA, EU, JP):*

- *самые популярные платформы ТОП-5, выясним и опишем различия в долях продаж,*

- *самые популярные жанры ТОП-5, выявим разницу и сделаем выводы,*

- *влияет ли рейтинг ESRB на продажи в отдельном регионе.*

*Создадим универсальный метод `top_by_region`, который создаст сводные таблицы и построит графики распреледения по количеству продаж в разрезе платформ, жанров и рейтинга ESRB в зависимости от региона.*

*Для прогноза продаж в будущем рассмотрим данные с 2013 года. Как технологии меняются очень быстро, так и вкусы людей вовлеченных в игру могут меняться быстро. Анализ предпочтений по регионам за 4 года позволит определить общие тенденции на текущий момент и планоровать на ближайщую перспективу актуальные потребности по регионам.*

```python
games_year_cut = games[(games['year_of_release'] > 2012)&(games['rating'] != 'without ESRB')]

def top_by_region(df, sales, title):    
    column_list = ['platform', 'genre', 'rating']    
    fig, axes = plt.subplots(1, 3, figsize=(25, 5))  
    ax = 0
    for element in column_list:        
        pivot = df.pivot_table(index = element,
                                            values = sales,
                                            aggfunc = 'sum').sort_values(by=sales,
                                                                         ascending=False).head(5)
        pivot.columns = ['sum' ]        
        pivot['percent'] = (pivot['sum']* 100 / pivot['sum'].sum()).round(2)  
                
        print(pivot)
        print('__________________________')        
        
        
        plt.grid(True)           
        sns.barplot(data=pivot, x=pivot.index, y='percent', ax=axes[ax])
        fig.suptitle(title, fontsize=20)
        ax += 1
    
    plt.show()
```

## **Проверка гипотез.**

<a id = 'hypothesis1'></a>

### **Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.**

**Гипотеза 1:Средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые.**

*Первая гипотеза двустороняя для средних двух генеральных совокупностей.*

*H0 (нулевая гипотеза) - средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые.*

*H1 (альтернативная гипотеза) - Средние пользовательские рейтинги платформ `Xbox One` и `PC` разные.*

*Создадим две независимые выборки пользовательских рейтингов платформ `Xbox One` и `PC` и сохраним их в переменных `xo_platform` и `pc_platform` соответствующие данные (актуальные данные с 2013 года).*

```python
xo_platform = games[(games['platform']=='XOne') & (games['year_of_release']>2012)]
pc_platform = games[(games['platform']=='PC') & (games['year_of_release']>2012)]
```

*Рассмотрим две независимые генеральные совокупности: `xb_platform` и `pc_platform`*

*Для проверки гипотезы будем использовать `t-критерий Стьюдента`, т.к. выборки независимы между собой.*

*Определим пороговое значение `alpha = 0.05`, означает 95-процентный уровень надежности.*

```python
xo_platform['user_score'].describe()
```

```python
pc_platform['user_score'].describe()
```

*Так как стандартные отклонения нельзя считать равными поэтому укажем аргумен `equal_var=False`*

*Ответ теста - "Отвергаем нулевую гипотезу" будет говорить о том, что гипотеза противоречит эмпирическим данным и скорей всего наше предположение неверно. Значит данные различаются.*

*Ответ теста - "Нет оснований отвергнуть нулевую гипотезу" значит, что предположение о равенстве среднего рейтинга верно.*

```python
import scipy.stats as st

print(f'Средний пользовательский рейтинг платформы Xbox One: {xo_platform.user_score.mean() :.2f}')
print(f'Средний пользовательский рейтинг платформы PC: {pc_platform.user_score.mean() :.2f}')

alpha = 0.05 # уровень статистической значимости.Хотим чтобы вероятность правильности нашего вывода была 95% и больше
print(f'Уровень статистической значимости: {alpha}')

# проведем стат. тест о равенстве средних значений времени двух выборок,
results = st.ttest_ind(xo_platform['user_score'],
                       pc_platform['user_score'],
                       equal_var=False)

print(f'Статистика t-теста: {results.statistic :.2f}')
print(f'Значение p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые')
else:
    print('Нет оснований отвергнуть нулевую гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и \
 PC одинаковые')
```

***Вывод***

*Полученное значение `p-value` (~0.206170) больше заданного уровня значимости 0.05, поэтому нет оснований отвергать нулевую гипотезу.*

*Есть основания говорить, что средние пользовательские рейтинги платформ `Xbox One` и `PC`  одинаковые.*

```python
import scipy.stats as st

print(f'Средний пользовательский рейтинг жанра Action: {action_genre.user_score.mean() :.2f}')
print(f'Средний пользовательский рейтинг жанра Sports: {sports_genre.user_score.mean() :.2f}')

alpha = 0.05 # уровень статистической значимости.Хотим чтобы вероятность правильности нашего вывода была 95% и больше
print(f'Уровень статистической значимости: {alpha}')

# проведем стат. тест о равенстве средних значений времени двух выборок,
results = st.ttest_ind(action_genre['user_score'],
                       sports_genre['user_score'],
                       equal_var=False)

print(f'Статистика t-теста: {results.statistic :.2f}')
print(f'Значение p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports одинаковые')
else:
    print('Нет оснований отвергнуть нулевую гипотезу о том, что средние пользовательские рейтинги жанров Action и\
 Sports одинаковые')
```

Цель это исследования определить критерии которые влияют на продажи компьютерных игр. Определение критериев позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Критериями , определяющими успешные продажи компьютерных игр, являются : 

- знание особенностей, предпочтений региона (жанр игры); 

- изветный производитель игр; 

- возможности и новые технологии платформ, на которых сделана игра; 

- целевая аудитория и все же отзывы.

Популярны компании - производители давно определились в жанрах, накопили аудиторию фанатов, которые ждут новые игры и  обязательно их преобретут.

Платформа - это техническая составляющая. Лучше платформа, более высокотехнологичная - значит качество и возможности игры будут лучше. Именно техническим прогрессом обусловлен жизненый цикл платформ. Возможно он будет сокращаться, так как мы постоянно наблюдаем скачки в развитии технологий.

Жанр - это то, что выбирает человек. Нельзя навязать свои пристрастия и приоритеты, нужно подстраиваться, быть многогранным, иметь продукт на любого пользователя. 

Более всего продаются игр для взрослых игроков, которым 17 лет и старше. Но в век высоких технологий, когда ребенок рождается с пультом в руке (а это потенциальные будущие клиенты и фанаты), нужно рынок насыщать качественной продукцией для детей.
